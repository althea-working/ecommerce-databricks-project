```python
import sys
import os
from datetime import datetime, timedelta

# =========================
# fix import path
# =========================
repo_path = os.path.abspath(os.path.join(os.getcwd(), "..", ".."))
sys.path.append(repo_path)

from src.gold.user_activity_metrics import build_user_activity_metrics
from src.common.logger import get_logger

logger = get_logger(__name__)

spark.conf.set("spark.sql.session.timeZone", "Asia/Shanghai")

# =========================
# params
# =========================
start_date = "2026-01-01"
end_date = "2026-04-08"

logger.info(f"Backfill from {start_date} to {end_date}")

# =========================
# run loop
# =========================
current = datetime.strptime(start_date, "%Y-%m-%d")
end = datetime.strptime(end_date, "%Y-%m-%d")

while current <= end:

    run_date_str = current.strftime("%Y-%m-%d")
    logger.info(f"Processing {run_date_str} computing user activity metrics")

    try:
        build_user_activity_metrics(spark, run_date_str)
    except Exception as e:
        logger.error(f"Failed for {run_date_str} computing user activity metrics: {e}", exc_info=True)
        raise

    current += timedelta(days=1)

logger.info("Backfill completed")
```
